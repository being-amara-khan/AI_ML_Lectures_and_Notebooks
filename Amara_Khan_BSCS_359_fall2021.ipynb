Make a scatterplot for each of the following:


1. People with age greater than 20 yrs and less than 60 yrs. (blue)
2. People with height greater than 6.5 ft. (yellow)
3. People with height less than 5.0 ft. (red)
4. People that fulfil condition 1. and do not fulfil condition 2. or 3. (green)

*Note: Display all the points on the same scatterplot in the colors specified. Use different shapes for each scatterplot as well.*





```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
age = np.random.randint(10, 71, size=100)
height = np.random.uniform(4.5, 7.2, size=100)


condition1 = (age > 20) & (age < 60)
condition2 = height > 6.5
condition3 = height < 5.0
condition4 = condition1 & (~condition2) & (~condition3)


plt.scatter(age[condition1], height[condition1], color='blue', label='Age (20-60)', marker='o')
plt.scatter(age[condition2], height[condition2], color='yellow', label='Height (>6.5 ft)', marker='s')
plt.scatter(age[condition3], height[condition3], color='red', label='Height (<5.0 ft)', marker='^')
plt.scatter(age[condition4], height[condition4], color='green', label='Age (20-60) & Not Height (>6.5 ft) & Not Height (<5.0 ft)', marker='x')


plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Scatterplot of Age vs Height for Different Conditions')
plt.legend()

# Show the plot
plt.show()

```
